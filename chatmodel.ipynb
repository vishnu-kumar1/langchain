```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough

text='ML ka matlab hai Machine Learning.Ye ek tareeka hai jisme hum computers ko sikhate hain ki woh khud se data se seekh kar kaam karein, bina har cheez ke liye alag se code likhe.Simple shabdon mein Socho jaise ek insaan experience se seekhta hai, waise hi machine data se seekhti hai.Hum kya karte hain? Hum machine ko bahut saara data dete hain (jise training data kehte hain).Machine kya karti hai?** Woh us data ko analyze karti hai, usmein patterns dhundhti hai, aur ek model banati hai.Phir kya hota hai? Ye model phir naye data par apply hota hai predictions ya decisions lene ke liye.'

llm1 = ChatGroq(
    model = 'llama-3.3-70b-versatile',
    temperature=0,
    api_key=''
)

llm2 = ChatGoogleGenerativeAI(
    model = "gemini-2.5-flash",
    temperature = 0,
    api_key = "AIzaSyBBSFoBzzOzMDi-FCrrKOGGkeYCwp4B6xI"
)

prompt = ChatPromptTemplate.from_messages([
    ('system','You are a helpfull assistance your work is to summarize the given text in englis with few lines'),
    ('human','this is the text {query}')
])
prompt2 = ChatPromptTemplate.from_messages([
    ('system','You are a helpfull assistance your work is to read the text on given topic and give a heading sugestion only 1 in english'),
    ('human','this is the text {query}')
])

model1 = prompt | llm1 | StrOutputParser() | {'query':RunnablePassthrough()} | prompt2 | llm2 | StrOutputParser()

response = model1.invoke({'query':text})
print(response)
```
